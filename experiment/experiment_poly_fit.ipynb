```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
n = 100
x = np.random.randn(n)
x.sort()
y = x * np.random.randn(n)

figure, ax = plt.subplots()
fit = np.polyfit(x, y, deg=1)
#print(fit)

fit2 = np.polyfit(x,y, deg=2)
#print(fit2)

fit3 = np.polyfit(x,y, deg =3)
#print(fit3)

fit4 = np.polyfit(x,y, deg =4)
#print(fit4)

fit5 = np.polyfit(x,y, deg =5)
#print(fit5)

fit50 = np.polyfit(x,y,deg=40)

fit100 = np.polyfit(x,y,deg=50)

ax.plot(x, fit[0] * x + fit[1], color='blue')
ax.plot(x, np.poly1d(fit2)(x), color='red')
ax.plot(x, np.poly1d(fit3)(x), color='yellow')
ax.plot(x, np.poly1d(fit4)(x), color='orange')
ax.plot(x, np.poly1d(fit5)(x), color='cyan')
ax.plot(x, np.poly1d(fit50)(x), color='magenta')
ax.plot(x, np.poly1d(fit100)(x), color='black')

ax.scatter(x, y)

figure.show()
```

### Polyfit

numpy polyfit function takes a set of points and, and tries to fir a polynomial of degree deg to best represent the set of points

It returns the cefficients of the n-degree polynomial that best fits the data.

for a degree = 1, the function returns two values a and b for the polynomial y = ax+b, therefore, we can plot it using the equation of a straight line

for more number of deegrees, we can either find the values of x^n , x^(n-1)...x^2, for all values in x and use the polynomial equaiton to generate plots.

The easier method is to use numpy's poly1d() method, which returns a function which is a polynomial defined by the coefficients that are passed. we can then use this method to generate the Y- axis values for our set of x values.

### Results

From the plot above, we see that as we increase the degree of the polynomial, it fits the sata better. We also see that there is very little difference between the graphs of a polynomial with degree 40 and that of degree 50. This is because, we have approximately found the polynomial that best fits the data, and trying to increase the degree will only result in zero or very small coefficient values.
